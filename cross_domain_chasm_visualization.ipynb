```python
import json
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
# ======= EDIT THESE WITH YOUR PUBMED METRICS =======

pubmed_metrics = {
    "ROUGE-1": 0.480,          # <-- replace
    "BERTScore_F1": 0.920,     # <-- replace
    "Redundancy": 0.050        # <-- replace
}

# arXiv metrics (already computed)
arxiv_metrics = {
    "ROUGE-1": 0.33702712477098173,
    "BERTScore_F1": 0.8180418610572815,
    "Redundancy": 0.02340549450549451
}
```

```python
metrics = list(pubmed_metrics.keys())

pubmed_values = [pubmed_metrics[m] for m in metrics]
arxiv_values = [arxiv_metrics[m] for m in metrics]

x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, pubmed_values, width, label="PubMed")
plt.bar(x + width/2, arxiv_values, width, label="arXiv")

plt.xticks(x, metrics)
plt.ylabel("Score")
plt.title("CHASM Cross-Domain Performance Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
percent_change = [
    ((arxiv_metrics[m] - pubmed_metrics[m]) / pubmed_metrics[m]) * 100
    for m in metrics
]

plt.figure(figsize=(8,5))
plt.bar(metrics, percent_change)

plt.axhline(0, color="black")
plt.ylabel("Percentage Change (%)")
plt.title("Relative Performance Change (arXiv vs PubMed)")
plt.tight_layout()
plt.show()

for m, p in zip(metrics, percent_change):
    print(f"{m}: {p:.2f}%")
```
